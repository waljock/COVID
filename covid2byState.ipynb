```python
import pandas as pd
import os
import seaborn as sb
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt


from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, Range1d, LabelSet, Label, Title
from bokeh.io import output_notebook
from bokeh.models.tools import HoverTool
```

[COVID Github Page](https://github.com/CSSEGISandData/COVID-19)

```python
url =  "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv"
url2 = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
#url3 = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv"

covx = pd.read_csv(url,index_col=1,parse_dates=[0])
covy = pd.read_csv(url2,index_col=0,parse_dates=[0])
#covz = pd.read_csv(url3,index_col=0,parse_dates=[0])
```

```python
cov0 = covx.reset_index()
cov0y = covy.reset_index()
#cov0z = covz.reset_index()
```

```python
cov0.drop(['iso2', 'UID', 'iso3', 'code3', 'FIPS', 'Admin2', 'Lat', 'Long_', 'Combined_Key'], inplace=True, axis=1)
```

```python
path = os.getcwd()
```

```python
cov0
```

```python
#change the file from wide to long
df0 = cov0.set_index(['Province_State','Country_Region']).stack().reset_index(level=2).reset_index()
```

```python
df0
```

```python
df = df0[df0[0]>=1]
```

```python
df
```

```python
#convert the date from text to date format
df['date'] = pd.to_datetime(df['level_2'])
df['value'] = df[0]
```

```python
df['date'].max()
```

```python
df
```

```python
#drop the province varible
df.drop(['level_2', 0, 'Country_Region'], inplace=True, axis=1)
```

```python
df
```

```python
#sum up country numbers
df2 = df.groupby(['Province_State', 'date']).sum().reset_index()
```

```python
sta = ['California', 'Texas', 'Florida']
```

```python
df3 = df2[(df2['Province_State'].isin(sta)) & (df2['date'] >= '2020-04-15')]
df3.sort_values(by=['Province_State', 'date'], inplace=True)
```

```python
df3.dtypes
```

```python
fig, ax = plt.subplots(figsize=(30, 15))
#plt.figure(figsize=(30, 15))


s = sb.lineplot('date', 'value', data=df3, hue='Province_State')

plt.title('Total COVID-19 Cases')
fig.autofmt_xdate()

# get current axis
#ax = plt.gca()
# get current xtick labels
#xticks = ax.get_xticks()
# convert all xtick labels to selected format from ms timestamp
#ax.set_xticklabels([pd.to_datetime(tm).strftime('%Y-%m-%d') for tm in xticks],
# rotation=50)

plt.show()
```

[Seaborn](https://seaborn.pydata.org/index.html)

```python
country = ['Florida']
```

```python
UScases = df2[(df2['Province_State'].isin(country))]
```

```python
YOUR_FONT_SIZE = '7pt'
#output_file("USCovid.html")

source = ColumnDataSource(UScases)

p = figure(x_axis_type="datetime", plot_width=750, plot_height=300)

p.circle(x='date', y='value', source=UScases, color = 'blue', size = 6)
p.line(x='date', y='value', line_width = 1, source=UScases, color = 'blue')
p.legend.location = 'bottom_right'

p.add_tools(HoverTool(
    tooltips=[
        ( 'date',   '@date{%F}'),
        ( '# of Cases', '@value'),
    ],

    formatters={
        'date'      : 'datetime', # use 'datetime' formatter for 'date' field
      },

    # display a tooltip whenever the cursor is vertically in line with a glyph
    mode='vline'
))

p.xaxis[0].axis_label = 'Date'
p.yaxis[0].axis_label = '# of Cases (cumulative)'


#p.add_layout(Title(text=market), 'above')
p.add_layout(Title(text=('Covid-19 Cases - Linear Scale' + '-State = ' + country[0])), 'above')

show(p)

```
